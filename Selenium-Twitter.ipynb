## Get List of people I am curreantly following on Twitter

```python
# Get Chromedriver
driver = webdriver.Chrome('/Users/ryanallred/Projects/selenium-twitter/chromedriver') 

# Launch URL
url = "https://twitter.com/ryanallred/following"
driver.get(url);
time.sleep(3) 

# Get input fields and button
username_input = driver.find_element_by_class_name('js-username-field')
password_input = driver.find_element_by_class_name('js-password-field')

# Login
username_input.send_keys(TWITTER_USERNAME)
password_input.send_keys(TWITTER_PASS)

submit_button = driver.find_element_by_xpath("//form/div/button[@type='submit']")
submit_button.click()
time.sleep(3) 

# Can we scroll down infinitely?
# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    
    # Random Scroll time between .5 and 1.5 seconds
    SCROLL_PAUSE_TIME = random.uniform(1.5, 2)
    
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# grab all of the user cards
cards = driver.find_elements_by_class_name('ProfileCard')

# Grab user information from cards
# Create list of named User tuples
from collections import namedtuple
User = namedtuple('User', ['username','follower'])

users = []

for card in cards:
    username = card.find_element_by_class_name('ProfileCard-screennameLink').text
    if card.find_elements_by_class_name('FollowStatus'):
        users.append(User(username, True))
    else:
        users.append(User(username, False))
```
